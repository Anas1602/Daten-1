```python
import pandas as pd

# Read the CSV file
datei_pfad = "C:/Users/anasl/Desktop/Daten 1/Daten/Output Data/Weather_Final_Output.csv"
daten = pd.read_csv(datei_pfad, sep=",", encoding='latin1')

# Extract unique Standort names
unique_standorte = daten['Standort'].unique()

# Display all unique names
print("All unique Standort names:")
for i, standort in enumerate(sorted(unique_standorte), 1):
    print(f"{i:2d}. {standort}")
```

```python
# Create mapping for corrupted German characters
zeichen_mapping = {
    'Ã¤': 'ä',  # a umlaut
    'Ã¶': 'ö',  # o umlaut  
    'Ã¼': 'ü',  # u umlaut
    'Ã„': 'Ä',  # A umlaut
    'Ã–': 'Ö',  # O umlaut
    'Ãœ': 'Ü',  # U umlaut
    'ÃŸ': 'ß',  # eszett
    'Ã‚': 'Â',  # A circumflex
    'Ã¡': 'á',  # a acute
    'Ã©': 'é',  # e acute
    'Ã­': 'í',  # i acute
    'Ã³': 'ó',  # o acute
    'Ãº': 'ú',  # u acute
    'Ã ': 'à',  # a grave
    'Ã¨': 'è',  # e grave
    'Ã¬': 'ì',  # i grave
    'Ã²': 'ò',  # o grave
    'Ã¹': 'ù',  # u grave
    'Ã¢': 'â',  # a circumflex
    'Ãª': 'ê',  # e circumflex
    'Ã®': 'î',  # i circumflex
    'Ã´': 'ô',  # o circumflex
    'Ã»': 'û'   # u circumflex
}

# Function to fix corrupted characters
# Fix encoding by re-decoding using UTF-8
def fix_encoding(text):
    if pd.isna(text):
        return text
    try:
        return text.encode('latin1').decode('utf-8')
    except:
        return text  # if decoding fails, return original


# Apply the fix to Standort column
daten['Standort'] = daten['Standort'].apply(fix_encoding)
```

```python
fixed_standorte = daten['Standort'].unique()

# Display a sample of remaining problematic names
problem_names = [name for name in sorted(fixed_standorte) if 'Ã' in name or '�' in name]
for name in problem_names[:20]:
    print(f"  {name}")

print(f"\nTotal unique Standorte: {len(fixed_standorte)}")
print(f"Names still with encoding issues: {len(problem_names)}")
```

```python
# Manual correction for remaining corrupted names
spezial_fixes = {
    'GroÃŸ': 'Groß',
    'MÃ¼nster': 'Münster',
    'OsnabrÃ¼ck': 'Osnabrück',
    'SchÃ¶nau am KÃ¶nigssee Bayern': 'Schönau am Königssee Bayern',
    'Ulm-MÃ¤hringen': 'Ulm-Mähringen'
}

def manuelle_korrektur(text):
    for falsch, richtig in spezial_fixes.items():
        text = text.replace(falsch, richtig)
    return text

daten['Standort'] = daten['Standort'].apply(manuelle_korrektur)
```
